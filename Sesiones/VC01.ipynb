### Complejidad cuadrática
---
La palabra in usada en una lista implica una búsqueda lineal. Anidar dos in, implica complejidad cuadrática del algoritmo.

```python
A = [1,2,6,7,12,13,15]
B = [2,3,4,7,13]

common = []

for num in A:
  if num in B:
    common.append(num)

print(common)
```

```python
r = [x for x in A if x in B]

print(r)
```

### Complejidad lineal
---
Emplear set() para complejidad lineal

```python
A = [1,2,6,7,12,13,15]
B = [2,3,4,7,13]

_A = set(A)
_B = set(B)

result = list(_A.intersection(_B))
```
